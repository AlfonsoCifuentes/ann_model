# Proyecto de Sitio Web Profesional para Modelo y Actriz

Genera un proyecto completo de sitio web profesional para modelo y actriz al nivel de portafolios internacionales, usando **Next.js 14 (App Router)** + **TypeScript** + **TailwindCSS** y desplegable en **Vercel**. Crea todo el código listo para ejecutar y da instrucciones de uso.

---

## Requisitos Detallados

### Branding y UX
- **Estilo:** Editorial/Fashion premium, minimalista y elegante. Mucho aire, tipografía de alta moda, microanimaciones sutiles.
- **Paleta (tokens Tailwind personalizados):**
    - `brand-bg: #0B0B0B` (fondo principal)
    - `brand-fg: #F5F5F5` (texto)
    - `brand-accent: #D9B06A` (dorado suave)
    - `brand-muted: #9A9A9A`
- **Tipografías:** Playfair Display (titulares) + Inter (texto). Cárgalas con `next/font/google` (preload).
- **Modo oscuro por defecto:** sobre fondo oscuro. Respeta `prefers-color-scheme`.

### Páginas y Secciones
- **/ Home**  
    Hero de impacto a pantalla completa con foto/loop video + claim, CTA “Book Me” y “View Portfolio”; highlights de 3–6 fotos tipo editorial; carrusel de clientes/logos; bloque “As seen in” con medios; video reel embebido; testimonios breves; CTA final.
- **/portfolio**  
    Grid masonry responsive con categorías (Editorial, Pasarela, Comercial, Acting Stills). Lightbox con teclado, swipe, zoom, y prefetch de imágenes siguiente/anterior.
- **/reel**  
    Página con video reel (mp4 y/o YouTube/Vimeo) con poster, controles custom accesibles, y transcripción breve.
- **/about**  
    Bio profesional (hitos, skills, medidas opcionales), retrato, botón a PDF compcard/press kit descargable.
- **/press**  
    Entradas tipo “case study / aparición” con fecha, medio, titular, imágenes y enlaces externos. Genera feed `/press/rss.xml`.
- **/booking**  
    Formulario (nombre, email, agencia/empresa, tipo de trabajo, presupuesto, fechas, mensaje) → usa Vercel Serverless API route y/o Formspree fallback. Validación Zod + react-hook-form, mensajes de éxito/error accesibles.
- **/contact**  
    Datos de contacto, redes sociales con iconos (Instagram, TikTok, IMDb, Spotlight, YouTube), mapa opcional.
- **Soporte i18n ES/EN**  
    Ejemplo: es por defecto, en secundaria. Selector en navbar y rutas localizadas.

### Contenido y Datos
- Usa un sistema de contenido en Markdown/MDX para portfolio (YAML frontmatter: `title`, `category`, `shoot_date`, `credits`, `cover`, `images[]`, `tags[]`) y press (YAML frontmatter: `title`, `outlet`, `date`, `cover`, `link`, `excerpt`, `gallery[]`).
- Crea 8–12 items de portfolio de ejemplo y 6 notas de prensa de ejemplo con imágenes de placeholder (usa `public/placeholders`).
- Genera un press kit/compcard PDF en `/public/press-kit.pdf` (placeholder y link en About).
- Crea un archivo `.env.example` con: `NEXT_PUBLIC_SITE_URL`, `NEXT_PUBLIC_ANALYTICS_DOMAIN` (Plausible), `FORMSPREE_ID` (si aplica).

### Componentes Clave
- **Navbar:** sticky con transparencia → sólido al hacer scroll.
- **Footer:** con enlaces rápidos y redes.
- **Hero:** con `next/image` (cover) o video autoPlay loop muted playsInline, degradado y headline tipográfico potente.
- **MasonryGrid:** accesible y performante (CSS columns o lib ligera sin dependencias pesadas).
- **Lightbox:** propio con portal, focus trap, ESC, flechas (teclado), swipe en móvil y preloading inteligente.
- **VideoPlayer:** con accesibilidad (captions VTT opcionales), poster, y controles visibles.
- Otros componentes: TestimonialCard, LogoMarquee (cinta infinita suave), TagPill, SectionHeader.
- **Extras:** LanguageSwitcher y Cookie/GDPR banner básico (solo analítica).

### Rendimiento, SEO y Accesibilidad
- **Core Web Vitals:**  
    - LCP < 2.5s  
    - CLS < 0.1  
    - Imágenes con `next/image` (responsive, lazy)  
    - Fuentes con display: swap
- **SEO:**  
    - `next/metadata` + OpenGraph/Twitter Cards  
    - sitemap.xml, robots.txt, canonical, alternate hreflang (ES/EN)  
    - schema.org Person y CreativeWork/Article para press
- **A11y:**  
    - Contraste AA, navegación por teclado, roles/aria, skip-links, focus visible
- **Analytics:**  
    - Integra Plausible (script defer y data-domain desde env)
- **RSS:** Genera `/press/rss.xml`.

### Infraestructura
- **Proyecto:** pnpm preferido. Tailwind configurado con tokens de marca y containers.
- **Comandos:**  
    - `pnpm create next-app@latest` (App Router, TS, ESLint, Tailwind).
    - Añade contentlayer o lectura MDX directa para contenido.
- **Scripts:** dev, build, start, lint, typecheck.
- **Despliegue:** Vercel (incluye `vercel.json` si hace falta). Soporte dominio personalizado. Instrucciones paso a paso para deploy.
- **Form Handling:** API route `/api/booking` + fallback Formspree si no hay backend.

### Archivos y Documentación
- Genera un `README.md` con guía de uso (añadir contenidos, traducir, añadir set nuevo al portfolio, configurar Plausible, variables de entorno, y despliegue).
